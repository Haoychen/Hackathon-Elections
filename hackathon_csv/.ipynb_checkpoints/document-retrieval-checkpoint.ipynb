# Load some text data - from wikipedia, pages on people

```python
twitter = graphlab.SFrame('Twitter_export.csv')
```

Data contains:  link to wikipedia article, name of person, text of article.

```python
twitter.head()
```

```python
len(twitter)
```

```python
twitter['word_count'] = graphlab.text_analytics.count_words(twitter['Title'])
```

```python
twitter.head()
```

# Compute TF-IDF for the corpus 

To give more weight to informative words, we weigh them by their TF-IDF scores.

```python
people['word_count'] = graphlab.text_analytics.count_words(people['text'])
people.head()
```

```python
tfidf = graphlab.text_analytics.tf_idf(twitter['word_count'])

# Earlier versions of GraphLab Create returned an SFrame rather than a single SArray
# This notebook was created using Graphlab Create version 1.7.1
if graphlab.version <= '1.6.1':
    tfidf = tfidf['docs']

tfidf
```

```python
twitter['tfidf'] = tfidf
twitter.head()
```

```python
tfidf_word = twitter[['tfidf']].stack('tfidf',new_column_name=['word','tfidf']).sort('tfidf',ascending=False)
```
